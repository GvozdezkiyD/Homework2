```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
file_path = "ex1.csv"
data = pd.read_csv(file_path)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö
print("üìä –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö:")
print(data.info())
print("\nüìã –ü–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ –¥–∞–Ω–Ω—ã—Ö:")
display(data.head())

# –ü—Ä–æ–≤–µ—Ä–∫–∞: –≤—Å–µ –ª–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ —á–∏—Å–ª–æ–≤—ã–µ
if data.select_dtypes(include=['float64', 'int64']).equals(data):
    print("\n‚úÖ –í—Å–µ —Å—Ç–æ–ª–±—Ü—ã —è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–æ–≤—ã–º–∏ –∏ –ø–æ–¥—Ö–æ–¥—è—Ç –¥–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞.")
else:
    print("\n‚ö†Ô∏è –ù–µ–∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–æ–ª–±—Ü—ã –Ω–µ —è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–æ–≤—ã–º–∏. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ–¥ –∞–Ω–∞–ª–∏–∑–æ–º.")
    raise ValueError("–î–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã.")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π
plt.figure(figsize=(14, 5))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è x
plt.subplot(1, 2, 1)
sns.histplot(data['x'], kde=True, bins=20, color='blue')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ x', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è x', fontsize=12)
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è y
plt.subplot(1, 2, 2)
sns.histplot(data['y'], kde=True, bins=20, color='green')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ y', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è y', fontsize=12)
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

plt.tight_layout()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ "—è—â–∏–∫–æ–≤ —Å —É—Å–∞–º–∏" –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, orient="h", palette="Set2")
plt.title('–Ø—â–∏–∫–∏ —Å —É—Å–∞–º–∏ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ x –∏ y', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['x'], y=data['y'], alpha=0.7, color='purple')
plt.title('–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è –º–µ–∂–¥—É x –∏ y', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.show()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞
corr_coeff, p_value = pearsonr(data['x'], data['y'])

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞
print("\nüîç –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞:")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞: {corr_coeff:.3f}")
print(f"P-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value:.3e}")

# –û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞
if abs(corr_coeff) > 0.7:
    print("\n‚úÖ –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–∏–ª—å–Ω–∞—è. –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –ü–∏—Ä—Å–æ–Ω–∞ –æ–ø—Ä–∞–≤–¥–∞–Ω–æ.")
else:
    print("\n‚ö†Ô∏è –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–ª–∞–±–∞—è. –†–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ –¥—Ä—É–≥–∏–µ –º–µ—Ç–æ–¥—ã –∞–Ω–∞–ª–∏–∑–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.")

print("\nüéØ –ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à–µ–Ω.")
```

```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
from scipy.stats import t

# –†–∞—Å—á–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞
corr_coeff, p_value = pearsonr(data['x'], data['y'])

# –ß–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
n = len(data)

# –†–∞—Å—á–µ—Ç t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
t_stat = corr_coeff * ((n - 2) ** 0.5) / ((1 - corr_coeff**2) ** 0.5)

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è t –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05
alpha = 0.05
t_critical = t.ppf(1 - alpha / 2, df=n - 2)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("üîç –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω–∞–ª–∏–∑–∞:")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞: {corr_coeff:.3f}")
print(f"P-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value:.3e}")
print(f"t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {t_stat:.3f}")
print(f"–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t (–ø—Ä–∏ Œ± = 0.05): ¬±{t_critical:.3f}")

# –í—ã–≤–æ–¥ –æ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
if abs(t_stat) > t_critical:
    print("\n‚úÖ –°–≤—è–∑—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–∞.")
else:
    print("\n‚ö†Ô∏è –°–≤—è–∑—å –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ —Å–ª–∞–±–∞—è –∏–ª–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç.")

# –í—ã–≤–æ–¥ –æ —Å—Ç–µ–ø–µ–Ω–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
if abs(corr_coeff) > 0.7:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–∏–ª—å–Ω–∞—è.")
elif abs(corr_coeff) > 0.3:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —É–º–µ—Ä–µ–Ω–Ω–∞—è.")
else:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–ª–∞–±–∞—è.")
```

```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import spearmanr

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞ 'ex2.csv'
file_path_ex2 = "ex2.csv"
data_ex2 = pd.read_csv(file_path_ex2)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö
print("üìä –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö:")
print(data_ex2.info())
print("\nüìã –ü–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ –¥–∞–Ω–Ω—ã—Ö:")
display(data_ex2.head())

# –ü—Ä–æ–≤–µ—Ä–∫–∞: –≤—Å–µ –ª–∏ –¥–∞–Ω–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ
if data_ex2.select_dtypes(include=['float64', 'int64']).equals(data_ex2):
    print("\n‚úÖ –í—Å–µ —Å—Ç–æ–ª–±—Ü—ã —è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–æ–≤—ã–º–∏.")
else:
    print("\n‚ö†Ô∏è –í –¥–∞–Ω–Ω—ã—Ö –æ–±–Ω–∞—Ä—É–∂–µ–Ω—ã –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –¥–∞–Ω–Ω—ã–µ –ø–µ—Ä–µ–¥ –∞–Ω–∞–ª–∏–∑–æ–º.")
    raise ValueError("–î–∞–Ω–Ω—ã–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –Ω–µ—á–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã.")

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(14, 5))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
plt.subplot(1, 2, 1)
sns.histplot(data_ex2.iloc[:, 0], kde=True, bins=20, color='blue')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–µ—Ä–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è –ø–µ—Ä–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞', fontsize=12)
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
plt.subplot(1, 2, 2)
sns.histplot(data_ex2.iloc[:, 1], kde=True, bins=20, color='green')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ç–æ—Ä–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è –≤—Ç–æ—Ä–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞', fontsize=12)
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

plt.tight_layout()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ "—è—â–∏–∫–æ–≤ —Å —É—Å–∞–º–∏" –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤
plt.figure(figsize=(10, 5))
sns.boxplot(data=data_ex2, orient="h", palette="Set2")
plt.title('–Ø—â–∏–∫–∏ —Å —É—Å–∞–º–∏ –¥–ª—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data_ex2.iloc[:, 0], y=data_ex2.iloc[:, 1], alpha=0.7, color='purple')
plt.title('–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏', fontsize=14)
plt.xlabel('–ü–µ—Ä–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫', fontsize=12)
plt.ylabel('–í—Ç–æ—Ä–æ–π –ø—Ä–∏–∑–Ω–∞–∫', fontsize=12)
plt.show()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞
spearman_corr, p_value_spearman = spearmanr(data_ex2.iloc[:, 0], data_ex2.iloc[:, 1])

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞
print("\nüîç –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –ø–æ –°–ø–∏—Ä–º–µ–Ω—É:")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞: {spearman_corr:.3f}")
print(f"P-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value_spearman:.3e}")

# –û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞
print("\nüìà –û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ:")
if spearman_corr > 0.7:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–∏–ª—å–Ω–∞—è, –∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –°–ø–∏—Ä–º–µ–Ω–∞ —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–∞.")
elif spearman_corr > 0.3:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —É–º–µ—Ä–µ–Ω–Ω–∞—è, –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞ –æ–ø—Ä–∞–≤–¥–∞–Ω–æ.")
else:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–ª–∞–±–∞—è, —Ä–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ –¥—Ä—É–≥–∏–µ –º–µ—Ç–æ–¥—ã –∞–Ω–∞–ª–∏–∑–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏.")
```

```python
from scipy.stats import spearmanr, t

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞
spearman_corr, p_value_spearman = spearmanr(data_ex2.iloc[:, 0], data_ex2.iloc[:, 1])

# –ß–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
n = len(data_ex2)

# –†–∞—Å—á–µ—Ç t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –°–ø–∏—Ä–º–µ–Ω–∞
t_stat_spearman = spearman_corr * ((n - 2) ** 0.5) / ((1 - spearman_corr**2) ** 0.5)

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è t –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05
alpha = 0.05
t_critical_spearman = t.ppf(1 - alpha / 2, df=n - 2)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("üîç –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω–∞–ª–∏–∑–∞ –ø–æ –°–ø–∏—Ä–º–µ–Ω—É:")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –°–ø–∏—Ä–º–µ–Ω–∞: {spearman_corr:.3f}")
print(f"P-–∑–Ω–∞—á–µ–Ω–∏–µ: {p_value_spearman:.3e}")
print(f"t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {t_stat_spearman:.3f}")
print(f"–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t (–ø—Ä–∏ Œ± = 0.05): ¬±{t_critical_spearman:.3f}")

# –í—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
if abs(t_stat_spearman) > t_critical_spearman:
    print("\n‚úÖ –°–≤—è–∑—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–∞.")
else:
    print("\n‚ö†Ô∏è –°–≤—è–∑—å –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞. –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –≤–µ–ª–∏—á–∏–Ω–∞–º–∏ —Å–ª–∞–±–∞—è –∏–ª–∏ –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç.")

# –í—ã–≤–æ–¥ –æ —Å—Ç–µ–ø–µ–Ω–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
print("\nüìà –í—ã–≤–æ–¥ –æ —Å—Ç–µ–ø–µ–Ω–∏ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:")
if abs(spearman_corr) > 0.7:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–∏–ª—å–Ω–∞—è.")
elif abs(spearman_corr) > 0.3:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —É–º–µ—Ä–µ–Ω–Ω–∞—è.")
else:
    print("–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è —Å–ª–∞–±–∞—è.")
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ —Å—Ä–µ–¥–Ω–µ–π –æ—à–∏–±–∫–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏
def mean_approximation_error(y_true, y_pred):
    """
    –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω—é—é –æ—à–∏–±–∫—É –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏.
    Args:
        y_true (array-like): –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
        y_pred (array-like): –ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
    Returns:
        float: –°—Ä–µ–¥–Ω—è—è –æ—à–∏–±–∫–∞ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö.
    """
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# –†–∞—Å—á–µ—Ç —Å—Ä–µ–¥–Ω–µ–π –æ—à–∏–±–∫–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏
approximation_error = mean_approximation_error(y, y_pred)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print(f"–°—Ä–µ–¥–Ω—è—è –æ—à–∏–±–∫–∞ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏: {approximation_error:.2f}%")
```

```python
from sklearn.metrics import mean_squared_error

# –†–∞—Å—á–µ—Ç F-–∫—Ä–∏—Ç–µ—Ä–∏—è
def calculate_f_statistic(y_true, y_pred, num_params):
    """
    –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç F-–∫—Ä–∏—Ç–µ—Ä–∏–π –¥–ª—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.
    Args:
        y_true (array-like): –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
        y_pred (array-like): –ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
        num_params (int): –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ (–≤–∫–ª—é—á–∞—è —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω).
    Returns:
        float: –ó–Ω–∞—á–µ–Ω–∏–µ F-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏.
    """
    n = len(y_true)  # –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
    rss = np.sum((y_true - y_pred) ** 2)  # –û—Å—Ç–∞—Ç–æ—á–Ω–∞—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
    tss = np.sum((y_true - np.mean(y_true)) ** 2)  # –û–±—â–∞—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
    ess = tss - rss  # –û–±—ä—è—Å–Ω–µ–Ω–Ω–∞—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤

    f_statistic = (ess / (num_params - 1)) / (rss / (n - num_params))
    return f_statistic

# –ß–∏—Å–ª–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –º–æ–¥–µ–ª–∏ (1 –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç + 1 —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω)
num_params = 2

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ F-–∫—Ä–∏—Ç–µ—Ä–∏—è
f_stat = calculate_f_statistic(y, y_pred, num_params)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print(f"F-–∫—Ä–∏—Ç–µ—Ä–∏–π: {f_stat:.3f}")

# –í—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
print("\nüìà –í—ã–≤–æ–¥:")
if f_stat > 4:  # –£—Å–ª–æ–≤–Ω—ã–π –ø–æ—Ä–æ–≥ –¥–ª—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, —É—Ç–æ—á–Ω–∏—Ç–µ –¥–ª—è –≤–∞—à–µ–≥–æ —Å–ª—É—á–∞—è
    print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –≤ —Ü–µ–ª–æ–º.")
else:
    print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–º.")
```

```python
from scipy.stats import t
import numpy as np

# –†–∞—Å—á–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –æ—à–∏–±–æ–∫ –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
def calculate_standard_errors(X, y, y_pred):
    """
    –í—ã—á–∏—Å–ª—è–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏ –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.
    Args:
        X (array-like): –ú–∞—Ç—Ä–∏—Ü–∞ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö (–¥–æ–±–∞–≤–ª–µ–Ω–∞ –∫–æ–ª–æ–Ω–∫–∞ —Å –µ–¥–∏–Ω–∏—Ü–∞–º–∏ –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞).
        y (array-like): –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
        y_pred (array-like): –ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
    Returns:
        tuple: —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏ –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞ –∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω–∞–∫–ª–æ–Ω–∞.
    """
    n = len(y)
    rss = np.sum((y - y_pred) ** 2)  # –û—Å—Ç–∞—Ç–æ—á–Ω–∞—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
    mse = rss / (n - 2)  # –°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (MSE)
    X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])  # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞ –¥–ª—è —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞
    XTX_inv = np.linalg.inv(X_with_intercept.T @ X_with_intercept)  # –û–±—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ (X'X)^-1
    std_errors = np.sqrt(np.diag(XTX_inv * mse))  # –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –æ—à–∏–±–∫–∏
    return std_errors

# –†–∞—Å—á–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –æ—à–∏–±–æ–∫
X_with_intercept = X  # –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
std_errors = calculate_standard_errors(X, y, y_pred)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
t_stat_intercept = intercept / std_errors[0]
t_stat_slope = slope / std_errors[1]

# –†–∞—Å—á–µ—Ç t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
corr_t_stat = pearson_corr * ((len(y) - 2) ** 0.5) / ((1 - pearson_corr**2) ** 0.5)

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è t –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞
alpha = 0.05
t_critical = t.ppf(1 - alpha / 2, df=len(y) - 2)

# –ó–∞–ø–∏—Å—å –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
confidence_interval_intercept = (intercept - t_critical * std_errors[0], intercept + t_critical * std_errors[0])
confidence_interval_slope = (slope - t_critical * std_errors[1], slope + t_critical * std_errors[1])

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("üìà –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω–∞–ª–∏–∑–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:")
print(f"–°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω (b0): {intercept:.3f}")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–∞–∫–ª–æ–Ω–∞ (b1): {slope:.3f}")
print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞: {std_errors[0]:.3f}")
print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω–∞–∫–ª–æ–Ω–∞: {std_errors[1]:.3f}")
print(f"t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ —á–ª–µ–Ω–∞: {t_stat_intercept:.3f}")
print(f"t-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –Ω–∞–∫–ª–æ–Ω–∞: {t_stat_slope:.3f}")
print(f"–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ t (–ø—Ä–∏ Œ± = 0.05): ¬±{t_critical:.3f}")
print("\nüìä –î–∏–∞–ø–∞–∑–æ–Ω—ã –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (95% –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª):")
print(f"–°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω: {confidence_interval_intercept}")
print(f"–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–∞–∫–ª–æ–Ω–∞: {confidence_interval_slope}")

# –í—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
print("\nüîç –í—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏:")
if abs(t_stat_intercept) > t_critical:
    print("–°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω (b0) —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")
else:
    print("–°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω (b0) –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")

if abs(t_stat_slope) > t_critical:
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–∞–∫–ª–æ–Ω–∞ (b1) —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")
else:
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–∞–∫–ª–æ–Ω–∞ (b1) –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")

if abs(corr_t_stat) > t_critical:
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")
else:
    print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –Ω–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º.")
```

```python
from scipy.stats import t
import numpy as np

# –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
def prediction_interval(x0, X, y, y_pred, intercept, slope, alpha=0.05):
    """
    –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è y –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º x0.
    Args:
        x0 (float): –ó–Ω–∞—á–µ–Ω–∏–µ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π (x), –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ –Ω—É–∂–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å y.
        X (array-like): –ù–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ.
        y (array-like): –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π.
        y_pred (array-like): –ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è.
        intercept (float): –°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω –º–æ–¥–µ–ª–∏.
        slope (float): –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–∞–∫–ª–æ–Ω–∞ –º–æ–¥–µ–ª–∏.
        alpha (float): –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –¥–ª—è –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞.
    Returns:
        tuple: –ü—Ä–æ–≥–Ω–æ–∑–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ y –∏ –µ–≥–æ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª.
    """
    n = len(y)
    rss = np.sum((y - y_pred) ** 2)  # –û—Å—Ç–∞—Ç–æ—á–Ω–∞—è —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤
    mse = rss / (n - 2)  # –°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞

    # –°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ x
    x_mean = np.mean(X)

    # –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –æ—à–∏–±–∫–∞ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    se_pred = np.sqrt(
        mse * (1 + (1 / n) + ((x0 - x_mean) ** 2) / np.sum((X - x_mean) ** 2))
    )

    # t-–∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
    t_critical = t.ppf(1 - alpha / 2, df=n - 2)

    # –ü—Ä–æ–≥–Ω–æ–∑–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ y
    y0_pred = intercept + slope * x0

    # –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
    lower_bound = y0_pred - t_critical * se_pred
    upper_bound = y0_pred + t_critical * se_pred

    return y0_pred, (lower_bound, upper_bound)

# –ó–Ω–∞—á–µ–Ω–∏–µ x0 –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
x0 = 6

# –ü—Ä–æ–≥–Ω–æ–∑–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
y0_pred, conf_interval = prediction_interval(x0, X.flatten(), y, y_pred, intercept, slope)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print("üîç –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –¥–ª—è x0 = 6:")
print(f"–ü—Ä–æ–≥–Ω–æ–∑–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ y: {y0_pred:.3f}")
print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª: ({conf_interval[0]:.3f}, {conf_interval[1]:.3f})")
```

```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞ 'ex4.csv'
file_path_ex4 = "/content/ex4.csv"
df4 = pd.read_csv(file_path_ex4)

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—Ç—Ä—É–∫—Ç—É—Ä—ã –¥–∞–Ω–Ω—ã—Ö
print("üìä –°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–∞–Ω–Ω—ã—Ö:")
print(df4.info())
print("\nüìã –ü–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ –¥–∞–Ω–Ω—ã—Ö:")
print(df4.head())

# –ò–º–µ–Ω–∞ —Å—Ç–æ–ª–±—Ü–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
x1, x2, y = df4.columns[0], df4.columns[1], df4.columns[2]

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º —Ä–∞—Å—Å–µ—è–Ω–∏—è
plt.figure(figsize=(18, 6))

# x1 vs y
plt.subplot(1, 3, 1)
plt.scatter(df4[x1], df4[y], alpha=0.7, color='blue')
plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è: {x1} vs {y}', fontsize=14)
plt.xlabel(x1, fontsize=12)
plt.ylabel(y, fontsize=12)
plt.grid(True)

# x2 vs y
plt.subplot(1, 3, 2)
plt.scatter(df4[x2], df4[y], alpha=0.7, color='green')
plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è: {x2} vs {y}', fontsize=14)
plt.xlabel(x2, fontsize=12)
plt.ylabel(y, fontsize=12)
plt.grid(True)

# x1 vs x2
plt.subplot(1, 3, 3)
plt.scatter(df4[x1], df4[x2], alpha=0.7, color='purple')
plt.title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è: {x1} vs {x2}', fontsize=14)
plt.xlabel(x1, fontsize=12)
plt.ylabel(x2, fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()

# –†–∞—Å—á–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
corr_x1_y, p_x1_y = pearsonr(df4[x1], df4[y])
corr_x2_y, p_x2_y = pearsonr(df4[x2], df4[y])
corr_x1_x2, p_x1_x2 = pearsonr(df4[x1], df4[x2])

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞
print("üîç –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:")
print(f"{x1} vs {y}: –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = {corr_x1_y:.3f}, P-–∑–Ω–∞—á–µ–Ω–∏–µ = {p_x1_y:.3e}")
print(f"{x2} vs {y}: –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = {corr_x2_y:.3f}, P-–∑–Ω–∞—á–µ–Ω–∏–µ = {p_x2_y:.3e}")
print(f"{x1} vs {x2}: –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = {corr_x1_x2:.3f}, P-–∑–Ω–∞—á–µ–Ω–∏–µ = {p_x1_x2:.3e}")

# –í—ã–≤–æ–¥ –æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
print("\nüìà –í—ã–≤–æ–¥—ã:")
if abs(corr_x1_x2) > 0.7:
    print(f"{x1} –∏ {x2} —Å–∏–ª—å–Ω–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç –º–µ–∂–¥—É —Å–æ–±–æ–π (–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = {corr_x1_x2:.3f}). –í–æ–∑–º–æ–∂–Ω–∞ –º—É–ª—å—Ç–∏–∫–æ–ª–ª–∏–Ω–µ–∞—Ä–Ω–æ—Å—Ç—å. –†–∞—Å—Å–º–æ—Ç—Ä–∏—Ç–µ –∏—Å–∫–ª—é—á–µ–Ω–∏–µ –æ–¥–Ω–æ–≥–æ –∏–∑ —Ñ–∞–∫—Ç–æ—Ä–æ–≤.")
else:
    print(f"{x1} –∏ {x2} –Ω–µ –∏–º–µ—é—Ç —Å–∏–ª—å–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç = {corr_x1_x2:.3f}). –ú–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞.")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import numpy as np

# –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df4[[x1, x2]])  # –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
y_scaled = scaler.fit_transform(df4[[y]])[:, 0]  # –°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –∑–∞–≤–∏—Å–∏–º–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model = LinearRegression()
model.fit(X_scaled, y_scaled)

# –ü–æ–ª—É—á–µ–Ω–∏–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
beta_coefficients = model.coef_

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("üìà –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–±–µ—Ç–∞-–∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã):")
print(f"{x1}: {beta_coefficients[0]:.3f}")
print(f"{x2}: {beta_coefficients[1]:.3f}")
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ (X) –∏ —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é (y)
X = df4[[x1, x2]].values  # –ù–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
y = df4[y].values  # –ó–∞–≤–∏—Å–∏–º–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model = LinearRegression()
model.fit(X, y)

# –ü–æ–ª—É—á–µ–Ω–∏–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ "—á–∏—Å—Ç–æ–π" —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
intercept_clean = model.intercept_  # –°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω
coefficients_clean = model.coef_  # –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

# –í—ã–≤–æ–¥ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
print("üìà –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã '—á–∏—Å—Ç–æ–π' —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:")
print(f"–°–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω (b0): {intercept_clean:.3f}")
print(f"{x1}: {coefficients_clean[0]:.3f}")
print(f"{x2}: {coefficients_clean[1]:.3f}")

# –ü—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π y
y_pred = model.predict(X)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –¥–∏–∞–≥—Ä–∞–º–º—ã —Ä–∞—Å—Å–µ—è–Ω–∏—è –ø—Ä–æ–≥–Ω–æ–∑–Ω—ã—Ö –∏ —Ä–µ–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred, alpha=0.7, color='blue', label='–ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linewidth=2, label='–ò–¥–µ–∞–ª—å–Ω–∞—è –ª–∏–Ω–∏—è')
plt.title('–ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ vs –†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è', fontsize=14)
plt.xlabel('–†–µ–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
plt.ylabel('–ü—Ä–æ–≥–Ω–æ–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score
from scipy.stats import f

# –†–∞—Å—á–µ—Ç –∏–Ω–¥–µ–∫—Å–∞ –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
r_squared = r2_score(y, y_pred)
multiple_correlation_index = r_squared ** 0.5  # –ò–Ω–¥–µ–∫—Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ (R)

# –ß–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
n = len(y)  # –ß–∏—Å–ª–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
k = X.shape[1] + 1  # –ß–∏—Å–ª–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (–≤–∫–ª—é—á–∞—è —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω)

# –†–∞—Å—á–µ—Ç F-–∫—Ä–∏—Ç–µ—Ä–∏—è
f_statistic = (r_squared / (k - 1)) / ((1 - r_squared) / (n - k))

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è F
alpha = 0.05  # –£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
f_critical = f.ppf(1 - alpha, dfn=k - 1, dfd=n - k)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("üìà –ò–Ω–¥–µ–∫—Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏:")
print(f"R (–ò–Ω–¥–µ–∫—Å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏): {multiple_correlation_index:.3f}")
print(f"R¬≤ (–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –¥–µ—Ç–µ—Ä–º–∏–Ω–∞—Ü–∏–∏): {r_squared:.3f}")

print("\nüîç F-–∫—Ä–∏—Ç–µ—Ä–∏–π:")
print(f"F-—Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞: {f_statistic:.3f}")
print(f"–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ F (–ø—Ä–∏ Œ± = {alpha}): {f_critical:.3f}")

# –í—ã–≤–æ–¥ –æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
print("\nüìä –í—ã–≤–æ–¥ –æ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è:")
if f_statistic > f_critical:
    print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ –≤ —Ü–µ–ª–æ–º.")
else:
    print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—ã–º.")
```
